actividad c


```python
# c1
df_limpio = df.copy()

# 1. Limpieza de texto y normalización
df_limpio['campus'] = df_limpio['campus'].str.strip().str.lower()
df_limpio['campus'] = df_limpio['campus'].replace({
    'slp cntro': 'slp centro',
    'slp centro ': 'slp centro'
})

df_limpio['modalidad'] = df_limpio['modalidad'].str.strip().str.lower()
df_limpio['modalidad'] = df_limpio['modalidad'].replace({'escolar': 'escolarizada'})

# 2. Eliminar duplicados
df_limpio = df_limpio.drop_duplicates(subset='id_estudiante', keep='first')

# 3. Corregir rangos (0–100) en promedio y asistencia
df_limpio['promedio'] = df_limpio['promedio'].clip(0, 100)
df_limpio['asistencia_pct'] = df_limpio['asistencia_pct'].clip(0, 100)

# 4. Corregir creditos_aprobados
df_limpio['creditos_aprobados'] = np.minimum(df_limpio['creditos_aprobados'], df_limpio['creditos_inscritos'])

# 5. Imputar faltantes
for col in ['edad', 'promedio', 'asistencia_pct']:
    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())
```

```python
print("Nulos por columna:\n", df_limpio.isnull().sum())

# Verificamos valores únicos de campus para asegurar normalización
print("\nCampus normalizados:", df_limpio['campus'].unique())

# Verificamos que los créditos aprobados no superen a los inscritos
errores_logicos = (df_limpio['creditos_aprobados'] > df_limpio['creditos_inscritos']).sum()
print(f"\nErrores de lógica en créditos: {errores_logicos}")
```

actividad d

```python
# d1
bajos_promedios = df_limpio[df_limpio['promedio'] < 80]
escolarizada_centro = df_limpio[(df_limpio['campus'] == 'slp centro') & (df_limpio['modalidad'] == 'escolarizada')]

print(f"Estudiantes con promedio < 80: {len(bajos_promedios)}")
print(f"Estudiantes en SLP Centro / Escolarizada: {len(escolarizada_centro)}")
```

```python
#d2
df_limpio['tasa_aprobacion'] = df_limpio['creditos_aprobados'] / df_limpio['creditos_inscritos']

#condiciones
condiciones = [
    (df_limpio['promedio'] < 70) | (df_limpio['asistencia_pct'] < 75),
    (df_limpio['promedio'] < 80)
]
opciones = ['Alto', 'Medio']

df_limpio['riesgo_academico'] = np.select(condiciones, opciones, default='Bajo')
```

actividad e

```python
# e1
df_final = pd.merge(df_limpio, apoyos, on='id_estudiante', how='left', indicator=True)
```

```python
# e2
print("--- Shape final ---")
print(df_final.shape)

print("\n--- Conteo de cruce (Merge) ---")
print(df_final['_merge'].value_counts())

print("\n--- Promedio por Campus ---")
print(df_final.groupby('campus')['promedio'].mean())

print("\n--- Distribución de Riesgo Académico ---")
print(df_final['riesgo_academico'].value_counts())
```

```python
df_limpio = df.copy()

# 9.1 Quitar espacios
df_limpio["campus"] = df_limpio["campus"].astype(str).str.strip()
df_limpio["modalidad"] = df_limpio["modalidad"].astype(str).str.strip()

# 9.2 Normalizar y corregir etiquetas
df_limpio["campus"] = df_limpio["campus"].str.lower().map({
    "slp centro": "SLP Centro",
    "slp cntro": "SLP Centro",
    "slp norte": "SLP Norte",
    "matehuala": "Matehuala"
}).fillna(df_limpio["campus"])

df_limpio["modalidad"] = df_limpio["modalidad"].str.lower().map({
    "escolarizada": "Escolarizada",
    "escolar": "Escolarizada",
    "mixta": "Mixta"
}).fillna(df_limpio["modalidad"])

# 9.3 Eliminar duplicados por id
df_limpio = df_limpio.drop_duplicates(subset=["id_estudiante"], keep="first").copy()

# 9.4 Corregir rangos
df_limpio["promedio"] = df_limpio["promedio"].clip(0, 100)
df_limpio["asistencia_pct"] = df_limpio["asistencia_pct"].clip(0, 100)

# 9.5 Corregir error lógico
df_limpio["creditos_aprobados"] = np.minimum(df_limpio["creditos_aprobados"], df_limpio["creditos_inscritos"])

# 9.6 Imputación simple con mediana
for col in ["edad", "promedio", "asistencia_pct"]:
    df_limpio[col] = df_limpio[col].fillna(df_limpio[col].median())

print("Limpieza completada ✅")
display(df_limpio)
```
